# Research the Money Flow

```python
import pyodbc
import pandas as pd

server='DEREK-PC\DEV2016'
db= 'Temp'
conn = pyodbc.connect('DRIVER={SQL Server};SERVER=' + server + ';DATABASE=' + db + ';Trusted_Connection=yes')

sqlStr = '''
with CTE as (
SELECT a.[Symbol], c.Sector
, a.[Date]
,b.Year
,convert(int, b.Month) Month
,[Close] * [Volume] MktValue
FROM [Temp].[dbo].[UsStocks] a
left join [BIDev].[dbo].[DimDate] b
on a.Date = b.Date
join [dbo].[Sectors] c
on a.Symbol = ltrim(rtrim(c.Symbol))
where a.Date > '2018-12-31'
)

select Year, Month, Sector, sum(MktValue) as MktValue
from CTE
group by Year, Month, Sector

'''

df1 = pd.io.sql.read_sql(sqlStr, conn)
df1.head(1)
```

```python
df2 = pd.pivot_table(df1, values = 'MktValue', index = ['Year', 'Month'], columns = 'Sector').reset_index()
```

```python
columnList = [col for col in df2.columns if col not in ['Year', 'Month']]
```

```python
import copy

df3 = df2.copy()
df3[columnList] = df3[columnList]/df3[columnList].iloc[0]
df3
```

```python
df4 = pd.melt(df3, id_vars = ['Year', 'Month'], var_name = 'Sector', value_name = 'RelativeVolume')
```

```python
df4.to_excel(r'RelativeVolumeBySector.xlsx', index=False)
```
